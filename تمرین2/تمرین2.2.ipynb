```python
import numpy as np
import matplotlib.pyplot as plt

# تابع فعال‌ساز سیگموئید
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# داده‌های آموزشی
X_train = np.array([[0.9], [0.75], [0.5], [0.65], [0.6], [0.35], [0.25]])
y_train = np.array([1, 1, 0.5, 0.5, 0.5, 0, 0])

# تنظیمات مدل
learning_rate = 0.1
epochs = 1000

# تعریف وزن‌ها و بایاس
np.random.seed(0)
w = np.random.randn(1)
b = np.random.randn(1)

# آموزش مدل
for epoch in range(epochs):
    # محاسبه خروجی مدل برای داده‌های آموزشی
    z = np.dot(X_train, w) + b
    y_pred = sigmoid(z)
    
    # محاسبه خطای مدل
    error = y_train - y_pred
    
    # به‌روزرسانی وزن‌ها و بایاس
    w += learning_rate * np.dot(X_train.T, error)
    b += learning_rate * np.sum(error)

# تعیین محدوده برای ترسیم منحنی
x_values = np.linspace(0, 1, 100)
y_values = sigmoid(np.dot(x_values[:, np.newaxis], w) + b)

# ترسیم داده‌ها و منحنی رگرسیون لجستیک
plt.scatter(X_train, y_train, color='blue', label='داده‌های آموزشی')
plt.plot(x_values, y_values, color='red', label='منحنی رگرسیون لجستیک')
plt.xlabel('مولفه اول (x)')
plt.ylabel('مولفه دوم (y)')
plt.title('رگرسیون لجستیک')

# مجموعه داده‌های تست
X_test = np.array([[0.15], [0.48], [0.59], [0.93], [1]])
for new_data in X_test:
    # دسته‌بندی داده جدید
    z_new = w * new_data + b
    y_pred_new = sigmoid(z_new)

    print("پیش‌بینی مدل برای داده جدید:", float(y_pred_new))

    if 0 <= y_pred_new < 0.4:
        predicted_class = 0
        print("عنوان: خوش‌خیم")
    elif 0.4 <= y_pred_new <= 0.7:
        predicted_class = 0.5
        print("عنوان: ازمایش مجدد")
    else:
        predicted_class = 1
        print("عنوان: بدخیم")

    print("کلاس داده جدید:", predicted_class)
    print()  # خط فاصله

    plt.scatter(new_data, predicted_class, color='green', label='داده جدید ({})'.format(float(new_data)))

plt.legend()
plt.show()
```
